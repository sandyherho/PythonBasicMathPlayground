# The unit circle

A circle that has always one unit from origin.

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from IPython.display import display,Math
%matplotlib inline
plt.style.use('ggplot')
```

```python
x = np.linspace(0,2*np.pi,100);

plt.plot(np.cos(x), np.sin(x), 'k');
plt.plot([-1.1,1.1],[0,0], '--', color=[.8,.8,.8]);
plt.plot([0,0],[-1.1,1.1], '--', color=[.8,.8,.8]);

angle = np.pi/4

plt.plot([0,np.cos(angle)],[0,np.sin(angle)], 'r');

plt.axis('square');
plt.axis([-1.1,1.1,-1.1,1.1]);
plt.xlabel('cos$(x)$');
plt.ylabel('sin$(x)$');
```

# Find a specific point on a Gaussian

Bell curve (normal curve)

### $$y = e^{\frac{-4ln(2)t^2}{h^2}}$$

### $$y = exp \large(\frac{-4 ln(2) t^2}{h^2})$$ 

```python
h = .999
t = np.linspace(-3,3,1001)

# gaussian
g = np.exp(-4*np.log(2)*(t**2)/(h**2))

# plot
plt.plot(t,g);
plt.xlabel('t');
plt.ylabel('y');
plt.title('Gaussian with FWHM ' + str(h));
```

```python
yval2find = .5
tvals = h*np.sqrt(np.log(yval2find)/(-4*np.log(2)))
print(tvals, -tvals)
```

## Exercise

```python
N = 100
G = np.zeros((N,len(t)))
h = np.zeros(N)

for i in range(N):
    h[i] = (i+1)/N
    G[i,:] = np.exp( -4 *np.log(2)*t**2 / h[i]**2)

plt.pcolormesh(t,h,G);
plt.xlabel('t');
plt.ylabel('y');
```

# Graphing the complex roots of unity

$$z^{n} = 1,  \quad z \quad \epsilon \quad C$$

$$ z = e^{2\pi i k / n} \text{ for } k = 0,1, \cdots , n-1$$

```python
n = 4

for k in range(n):
    z = sym.exp(2*sym.pi*sym.I*k/n)
    display(Math('(%s)^{%s} \\Rightarrow %s'%(sym.latex(z),n, sym.latex(z**n))))
```

```python
n = 7
for k in range(n):
    z = np.exp(2*np.pi*1j*k/n)
    plt.plot([0,np.real(z)],[0,np.imag(z)]);

x = np.linspace(0,2*np.pi, 100)
plt.plot(np.cos(x), np.sin(x), color='grey');
plt.axis('square');
plt.show();
```

# Log-spaced and linearly spaced numbers

linear : 1,2,3,4,5

log-space: 1, 1.5, 2.2, 3.3, 5

```python
np.linspace(1,2,10)
```

```python
np.logspace(1,2,10) # 10^n
```

```python
np.logspace(np.log10(1), np.log10(2), 10)
```
