# Graphing circles

$$\left(y - k\right)^{2} + \left(x - h\right)^{2} = r^{2}$$

```python
h = 2
k = -3
r = 3

axlim = r + np.max((abs(h),abs(k)))
x = np.linspace(-axlim,axlim,100)
X,Y = np.meshgrid(x,x)

Fxy = (X-h)**2 + (Y-k)**2 - r**2

plt.contour(X,Y,Fxy,0);
plt.plot([-axlim, axlim],[0,0], 'k--');
plt.plot([0,0],[-axlim, axlim], 'k--');
plt.plot(h,k,'go');
plt.gca().set_aspect('equal');
```

# Graphing ellipses

$$\frac{(y - k)^{2}}{b^{2}} + \frac{(x - h)^{2}}{a^{2}} - 1 = 0$$

Center: $\left(h, k\right)$

Radii: $a, b$

```python
h = 1
k = 2
a = 2
b = 3

axlim = np.max((a,b)) + np.max((abs(h),abs(k)))
x = np.linspace(-axlim,axlim,100)
y = np.linspace(-axlim,axlim,100)
X,Y = np.meshgrid(x,y)

# create the function
ellipse = (X-h)**2/a**2 + (Y-k)**2/b**2 - 1

plt.plot(h,k,'go');
plt.contour(X,Y,ellipse, 0);
plt.gca().set_aspect('equal');
```

# Graphing hyperbolas

$$\frac{(y-k)^2}{b^2} - \frac{(x-h)^2}{a^2} - 1 = 0$$

Center: $(h,k)$

Radii: $a,b$

```python
a = 1
b = .5
h = 1
k = 2
axlim = 2*np.max((a,b)) + np.max((abs(h), abs(k)))
x = np.linspace(-axlim, axlim, 100)

X,Y = np.meshgrid(x,x);

# create function
hyperbola = (X-h)**2/a**2 - (Y-k)**2/b**2 - 1
plt.contour(X,Y,hyperbola,0);
plt.plot(h,k,'go');
plt.plot([-axlim,axlim], [0,0], 'w--');
plt.plot([0,0], [-axlim, axlim], '--', color =[1,1,1]);

plt.gca().set_aspect('equal');
```
