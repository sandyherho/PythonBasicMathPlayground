```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
%matplotlib inline
```

# Introduction to random numbers

```python
nums = np.random.rand(1000) # uniform distribution 0 -1
plt.plot(nums, 's');
```

```python
plt.hist(nums);
```

```python
nums = np.random.rand(1000)
nums += 3 # mulai di angka 3 hingga 4

plt.plot(nums, 's');
plt.show()

plt.hist(nums);
plt.show()
```

```python
nums = np.random.rand(1000)

minval = 2
maxval = 17

nums = nums* (maxval - minval) + minval

plt.plot(nums, 's');
plt.show()

plt.hist(nums);
plt.show()
```

```python
nums = np.random.randn(1000) # distribusi normal (center = 0, std = 1)

plt.plot(nums, 's', alpha = .4);
plt.show()

plt.hist(nums);
plt.show()
```

## Exercises

Generate a normal distribution mean = 15, std = 4.3

```python
desired_mean = 15
desired_std = 4.3

nums = np.random.randn(1000)
nums = nums - np.mean(nums) # getting empirical means (normalisasi mean = 0)
nums = nums/np.std(nums) # getting empirical std (normalisasi std = 1)

nums = nums*desired_std + desired_mean

plt.subplot(121); #(1 row, 2 cols, 1st subplot)
plt.plot(nums, 's', alpha=0.5);

plt.subplot(122); 
plt.hist(nums, alpha = .5);

plt.show()

print('This distribution has a mean of %s and a standard-deviation of %s'%(np.mean(nums), np.std(nums)))
```
