## Exercises

Generate a normal distribution mean = 15, std = 4.3

```python
desired_mean = 15
desired_std = 4.3

nums = np.random.randn(1000)
nums = nums - np.mean(nums) # getting empirical means (normalisasi mean = 0)
nums = nums/np.std(nums) # getting empirical std (normalisasi std = 1)

nums = nums*desired_std + desired_mean

plt.subplot(121); #(1 row, 2 cols, 1st subplot)
plt.plot(nums, 's', alpha=0.5);

plt.subplot(122); 
plt.hist(nums, alpha = .5);

plt.show()

print('This distribution has a mean of %s and a standard-deviation of %s'%(np.mean(nums), np.std(nums)))
```

Plot unit vectors with random phase angles. What random disributions are most sensible?

```python
angs = np.random.rand(500)*2*np.pi

for i in angs:
    plt.plot([0, np.cos(i)], [0, np.sin(i)]);

plt.axis('square')
plt.axis([-1, 1, -1, 1])
plt.axis('off')
plt.show()
```

# Converting between radians and degrees

$$2\pi = 360 ^{\circ}$$

$$rad. = \frac{360 ^{\circ}}{2\pi} = \frac{180 ^{\circ}}{\pi}$$

$$deg. = \frac{2\pi}{360 ^{\circ}} = \frac{\pi}{180 ^{\circ}}$$

```python
deg = 6876546546876

rad = deg*np.pi/180

rad = rad %(2*np.pi) #( between 0 to 2*pi)

print("%g degrees is %g radians" %(deg, rad))
```

```python
rad = 4*np.pi

deg = (180 * rad / np.pi) % 360 #(between 0 to 360)
deg
```

Is it hard? Use numpy built-in function!!!

```python
deg = 720
rad = np.deg2rad(deg) % (2*np.pi)
rad
```

```python
rad = 4*np.pi

deg = np.rad2deg(rad) % 360
deg
```

```python
def angle_convert_plot():
    
    import numpy as np
    
    # get the input from the user
    angle = eval(input('What angle to convert? '))
    unit = input('Which unit did you input? (radian or degree) ')
    
    if unit[0].lower()=='r':
        rad = angle
        deg = np.rad2deg(angle)
    elif unit[0].lower()=='d':
        deg = angle
        rad = np.deg2rad(angle)
    else:
        raise ValueError('Unknown unit!')
    
    # convert to < 360
    deg = deg % 360
    rad =rad % 2*np.pi
    
    # make the graph
    plt.plot([0,1],[0,0], 'r', lw = 2);
    plt.plot([0, np.cos(rad)], [0, np.sin(rad)], 'r', lw=3);
    plt.axis('square');
    plt.axis([-1,1, -1, 1]);
    plt.grid();
    plt.title('Angle is %s$^o$ or %s radian' %(deg, rad))
```

# The Pythagorean theorem

```python
a = 3
b = 4
c = np.sqrt(a**2 + b**2)

# draw the triangle
plt.plot([0,a], [0,0], 'k', lw = 2);
plt.plot([0,0], [0,b], 'k');
plt.plot([0,a], [b,0], 'k');

plt.plot([.3,.3], [0,.3], 'k',lw=1);
plt.plot([0,.3],[0.3,.3], 'k', lw=1);

plt.text(a/3,-.3,'a = ' + str(a), fontsize=15);
plt.text(.1, b/3, 'b = ' + str(b), fontsize=15);
plt.text(a/2 + .1, b/2, 'c = ' + str(c), fontsize=15);

plt.axis('square');
axlim = np.max((a,b)) + .5
plt.axis([-.5, axlim, -.5, axlim]);
plt.axis('off');
```

## Exercises

 $$ y_{1} = \sin(\theta + \cos(\theta)) $$
 
 $$ y_{2} = \cos(\theta + \sin(\theta))$$

```python
# 1

thetas = np.linspace(0, 4*np.pi, 1000)
y1 = np.sin(thetas + np.cos(thetas))
y2 = np.cos(thetas + np.sin(thetas))

plt.plot(thetas, y1, 'r', lw = 2, label = '$\\sin(\\theta + \\cos(\\theta))$');
plt.plot(thetas, y2, 'b', lw = 2, label = '$\\cos(\\theta + \\sin(\\theta))$');
plt.legend();
```

```python
# 2

thesym = sym.pi * 3 / 16
theta = float(thesym)

# plotting the unit circle
x = np.linspace(0, 2*np.pi, 200)
plt.plot(np.cos(x), np.sin(x), color = 'gray');
plt.plot([-1.1,1.1], [0,0], '--', color='grey');
plt.plot([0,0],[-1.1,1.1], '--', color='grey');

# plot the arc
plt.plot(np.cos(np.linspace(0,theta)), np.sin(np.linspace(0, theta)), 'k', lw = 3);

# plot vector
plt.plot([0,np.cos(theta)], [0, np.sin(theta)], 'k:');
plt.plot(np.cos(theta), np.sin(theta), 'ko');

# plot separate sine/cosine components
plt.plot([0,0], [0,np.sin(theta)], 'r');
plt.plot([0, np.cos(theta)], [0,0], 'b');

plt.plot([0,np.cos(theta)], [np.sin(theta), np.sin(theta)], 'r:');
plt.plot([np.cos(theta), np.cos(theta)], [0,np.sin(theta)], 'b:');


plt.title('$\\theta = %s$' %(sym.latex(thesym)));
plt.gca().set_aspect('equal');
```
