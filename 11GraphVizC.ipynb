# Making images from matrices

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
%matplotlib inline
```

```python
A = [[1,2], [1,4]]
print(A)
```

```python
plt.imshow(A);
plt.xticks([0,1]);
plt.yticks([0,1]);
```

$$a_{i,j} = 3i - 4j$$

```python
A = np.zeros((10,14))
print(np.shape(A))

for i in range(0,np.shape(A)[0]):
    for j in range(0, np.shape(A)[1]):
        A[i,j] = 3*i - 4*j
print(A)
```

```python
plt.imshow(A);
for i in range(0,np.shape(A)[0]):
    for j in range(0, np.shape(A)[1]):
        plt.text(j,i,int(A[i,j]), horizontalalignment='center', verticalalignment='center');
        # dituker j,i karena di dalam plotting urutan antara kolom (y) dan baris (x) beda dgn matriks
        # matriks baris (i-x), kolom(j-y)
plt.set_cmap('coolwarm_r');
```

## Exercise

$$c_{i,j} = (-1)^{i + j}$$

```python
n = 10
C = np.zeros((n,n))

for i in range(0,n):
    for j in range(0,n):
        C[i,j] = (-1)**(i + j)

print(C)
```

```python
plt.imshow(C);
plt.set_cmap('gray')
plt.tick_params(labelleft=False, labelbottom=False)
```

# Drawing patches with polygons

```python
from matplotlib.patches import Polygon
```

```python
y = np.array([[1,1],[2,3],[3,1]]) # harus kaya gini kalau mau gambar polygon pake array x,y
p = Polygon(y, facecolor='m', alpha=.3)

y1 = np.array([[2,2],[2.5,4],[3.5,1]]) 
p1 = Polygon(y1, facecolor='b', alpha=.3, edgecolor='k')

fig,ax = plt.subplots();
ax.add_patch(p);
ax.add_patch(p1);
ax.set_xlim([0,4]);
ax.set_ylim([0,4]);
```

## Exercise

$$f(x) = -x^{2}$$

```python
x = np.linspace(-2,2,101)
f = -x**2

pnts = np.vstack((x,f)).T

p = Polygon(pnts, facecolor='g',alpha=.2)

y = np.array([[-.5,-4], [-.5,-2.5], [.5,-2.5], [.5, -4]])
p1 = Polygon(y, facecolor='k')

fig,ax = plt.subplots();
ax.add_patch(p)
ax.add_patch(p1)

plt.plot(x,f, 'k')
plt.plot([-2,2],[-4,-4], 'k');
plt.axis('off');
plt.savefig("iglo.pdf")
```
