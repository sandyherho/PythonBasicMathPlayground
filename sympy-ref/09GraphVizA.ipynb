# Coordinates on a plane

```python
import matplotlib.pyplot as plt
import sympy as sym
%matplotlib inline
#plt.style.use('ggplot')
```

```python
x = 3
y = 5

plt.plot(x,y, 'ro')
plt.axis('square')
plt.axis([-6,6,-6,6]) #[-x,x,-y,y]
plt.grid()
```

```python
plt.plot(x,y, 'gs')
plt.axis('square')
plt.axis([-6,6,-6,6]) #[-x,x,-y,y]
plt.grid()
```

```python
x = [-4,5,2,4,5,6,3]
y = [0,3,4,5,2,-3,4]

for i in range(0,len(x)):
    plt.plot(x[i],y[i], 'o', label = 'point %s' %(1 + i));
plt.axis('square');
plt.grid();
plt.legend();
```

```python
plt.plot(4,2,'r+');
axis = plt.gca()

ylim = axis.get_ylim()
print(ylim)
axis.set_ylim([ylim[0], 5.4])
```

## Exercise

```python
x = sym.symbols("x")
y = x**2 - 3*x 

for pnti in range(-11,12):
    plt.plot(pnti, y.subs(x,pnti), 'o');
plt.xlabel('x');
plt.ylabel('f(x) = %s' %y);
```

# Graphing lines: start-end

```python
p1 = [-3,1]
p2 = [4,4]

plt.plot([p1[0],p2[0]],[p1[1], p2[1]], color=[1,.2,.7], linewidth=5); #RGB

plt.axis("square")
plt.axis([-6,6,-6,6])
```

```python
x = 3
y = 5

plt.plot(x,y,'ro');
plt.plot([0,x],[0,y],'r');

plt.axis("square")
plt.axis([-6,6,-6,6])
plt.plot([-6,6],[0,0],'k')
plt.plot([0,0],[-6,6],'k')

plt.grid()
```

# Graphing lines : slope-intercept

$y = mx + b$

```python
import numpy as np
x = [-5,5]
m = 0.3
b = 0.1

#y = [340, 2345]
#for i in range(0,len(x)):
#    y[i] = m*x[i] + b
y = m*np.array(x) + b
print(x,y)
plt.plot(x,y, label='y = %sx + %s'%(m,b));

plt.axis('square');
plt.grid();

plt.xlim(x);
plt.ylim(x);

axis = plt.gca();
plt.plot(axis.get_xlim(), [0,0], 'k--');
plt.plot([0,0], axis.get_xlim(), 'k--');

plt.legend();
```
