# Coordinates on a plane

```python
import matplotlib.pyplot as plt
%matplotlib inline
#plt.style.use('ggplot')
```

```python
x = 3
y = 5

plt.plot(x,y, 'ro')
plt.axis('square')
plt.axis([-6,6,-6,6]) #[-x,x,-y,y]
plt.grid()
```

```python
plt.plot(x,y, 'gs')
plt.axis('square')
plt.axis([-6,6,-6,6]) #[-x,x,-y,y]
plt.grid()
```

```python
x = [-4,5,2,4,5,6,3]
y = [0,3,4,5,2,-3,4]

for i in range(0,len(x)):
    plt.plot(x[i],y[i], 'o', label = 'point %s' %(1 + i));
plt.axis('square');
plt.grid();
plt.legend();
```

```python
plt.plot(4,2,'r+');
axis = plt.gca()

ylim = axis.get_ylim()
print(ylim)
axis.set_ylim([ylim[0], 5.4])
```
