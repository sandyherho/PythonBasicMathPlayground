# Graphing rational functions

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-3,4)
y = np.zeros(len(x))

for i in range(0,len(x)):
    y[i] = 2 - x[i]**2

plt.plot(x,y, 's-');
```

```python
x = np.linspace(-3,4,100)
y1 = 2 + abs(x)**(1/3)
y2 = 2 + abs(x)**(1/2)

plt.plot(x,y1, 'm-');
plt.plot(x,y2, 'k-');
```

## Exercise

```python
x = np.linspace(-4,4,200)
e = range(-1,4)

for i in e:
    y = x**i
    plt.plot(x,y, label="$y = x^{%s}$" %i, linewidth=4);


plt.legend();
plt.xlim([x[0], x[-1]]);
plt.ylim([-20,20]);
```

# Plotting with sympy

```python
import sympy as sym
import sympy.plotting.plot as symplot
```

```python
from sympy.abc import x

y = x**2

p = symplot(y, show = False);
p.xlim = [0,50]
p[0].line_color = 'm'
p.title = 'This is a nice plot'
p.show();
```

```python
print(p)
```

```python
x, a = sym.symbols('x,a')

expr = a/x
n = 1
p = symplot(expr.subs(a,n), (x,-5,5), show=False)

p[0].label = '$y=%s$' %sym.latex(expr.subs(a,n))

p.extend(symplot(expr.subs(a,2), show=False))

p[1].label = '$y=%s$' %sym.latex(expr.subs(a,2))
p[0].line_color = 'r'
p.ylim = [-5,5]

p.legend = True
p.show();
```
